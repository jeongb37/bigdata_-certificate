```python
import pandas as pd

data = pd.read_csv("../input/bigdataexam/t1-data1.csv")
print(data)
```

```python
# 특정 컬럼 각 도수 구하기
print(data['city'].value_counts())
```

```python
print(data['f4'].unique())
```

```python
data1 = data.copy()
# 최빈값
data1['city'].mode()
```

```python
# 큰 값부터 정렬 10개 뽑기
data1_1 = data1.sort_values('f5', ascending =False)
print(data1_1)
print(data1_1['f5'][:10])
```

```python
# 중앙값으로 변경 후 f5의 분산, 표준편차 구하기 
med = data1_1['f5'][:10].median()
print(med)
data1_1['f5'][:10] = med
print(data1_1)
var = data1_1['f5'].var()
std = data1_1['f5'].std()
print(var, std)
```

```python
# 조건부 값 변경
# loc와 iloc 적용하기!!!!!!!
data1_1.loc[(data1_1['f4']=='ISFP')|(data1_1['f4']=='ENFP'), 'f4'] = 'ESFP'
print(data1_1)
```

```python
# 조건부 뽑기
print(data1_1[(data1_1['city']=='서울')&(data1_1['f4']=='ESFP')]['f5'].max())
```

### 결측치 처리

```python
# 결측치 확인
data1.isnull().sum()
```

```python
# 데이터의 50%만 사용해서 결측치 처리
v50 = int(len(data1)*0.5)
data1_2 = data1[:v50]
# print(data1_2)
data1_2['f1'] = data1_2['f1'].fillna(data1_2['f1'].mean())
print(data1_2.head())
```

```python
# 5번째 수로 결측치 처리
data1_3 = data1.copy()
five = data1_3.loc[4, 'f1']
print(five)
print(data1_3.isnull().sum())
```

```python
data1_3['f1'] = data1_3['f1'].fillna(five)
print(data1_3.isnull().sum())
```

### 이상치

#### IQR

```python
data2 = data.copy()
print(data2.head())
```

```python
data2.describe()
```

```python
# IQR
Q3 = data2['f5'].quantile(.75)
Q1 = data2['f5'].quantile(.25)
IQR = Q3-Q1
print(Q3, Q1, IQR)
```

```python
min_out = Q1 - 1.5*IQR
max_out = Q3 + 1.5*IQR
print(min_out, max_out)

print(data2[(data2['f5']<min_out)&(data2['f5']>max_out)])
```
